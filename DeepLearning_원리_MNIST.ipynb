 2. 활성화 함수 : 전이 함수(transfer function) 또는 비선형성(nonlinearity)
 - 활성화 함수가 가중합의 선형 결합을 비선형 모델로 만듦
 - 각 퍼셋트로 내 계산의 마지막 단계에 배치되어 뉴런의 발화 여부를 결정함.
 - 가중합을 계산해서 그대로 출력해도 되는데 굳이 왜 할성화 함수를 사용하는가? 신경망에 비선형성을 도입하기 위해서
 - 비선형성이 없이는 아무리 층수가 많은 다층 퍼셉트론도 단일 퍼셉트론과 동일함.
 - 활성화 함수는 출력값을 특정 구간 안으로 제약하는 효과도 있음.  






```python
import numpy as np
import matplotlib.pyplot as plt

# 활성화 함수 정의
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, x * alpha)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

# 입력 값 생성
x = np.linspace(-10, 10, 100)

# 활성화 함수별 출력 값 계산
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)

# Softmax는 여러 입력 값에 대해 계산하도록 설정
x_softmax = np.linspace(-10, 10, 100)
# 다차원 배열을 만들어 소프트맥스 함수의 출력 분포를 시각화
X_softmax = np.vstack([x_softmax, x_softmax + 1, x_softmax + 2])
y_softmax = softmax(X_softmax)

# 활성화 함수 그래프 그리기
plt.figure(figsize=(14, 10))

plt.subplot(2, 3, 1)
plt.plot(x, y_sigmoid, label="Sigmoid")
plt.title("Sigmoid")
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(x, y_tanh, label="Tanh")
plt.title("Tanh")
plt.grid(True)

plt.subplot(2, 3, 3)
plt.plot(x, y_relu, label="ReLU")
plt.title("ReLU")
plt.grid(True)

plt.subplot(2, 3, 4)
plt.plot(x, y_leaky_relu, label="Leaky ReLU")
plt.title("Leaky ReLU")
plt.grid(True)

plt.subplot(2, 3, 5)
for i in range(X_softmax.shape[0]):
    plt.plot(x_softmax, y_softmax[i], label=f"Softmax output {i+1}")
plt.title("Softmax")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf

class NaiveDense:
    # 층을 정의하는 클래스
    def __init__(self, input_size, output_size, activation):
        # input_size : 입력 특성의 수
        # output_size : 출력 특성의 수
        # activation : 활성화 함수
        self.activation = activation

        # 입력, 출력 수에 따라 가중치 행렬 크기 생성
        w_shape = (input_size, output_size)
        # 균일 분포로 가중치 초기화
        w_initial_value = tf.random.uniform(w_shape, minval=0, maxval=1e-1)
        self.W = tf.Variable(w_initial_value)

        # 출력 특성 수에 따른 편향 초기화
        b_shape = (output_size,)
        b_initial_value = tf.zeros(b_shape)
        self.b = tf.Variable(b_initial_value)

    def __call__(self, inputs):
        # 인스턴스 호출 시 입력 데이터를 처리하고 결과 반환
        return self.activation(tf.matmul(inputs, self.W) + self.b)

    @property
    def weights(self):
        # 가중치와 편향을 리스트로 반환
        return [self.W, self.b]

# NaiveDense 클래스 테스트
testNaiveDense = NaiveDense(10, 10, activation=tf.nn.relu)
input_data = tf.random.normal((1, 10))  # 입력 데이터 예시, 크기는 (배치 크기, 입력 특성 수)여야 함
output = testNaiveDense(input_data)

print("testNaiveDense.W : ", testNaiveDense.W)
print("testNaiveDense.b : ", testNaiveDense.b)
print("output :", output)
print("testNaiveDense.weights : ", testNaiveDense.weights)
```

```python
class NaiveSequential:
    def __init__(self, layers):
        self.layers = layers  # 층들의 리스트

    def call(self, inputs):
        x = inputs
        for layer in self.layers:
            x = layer(x)  # 각 층에 입력을 전달하고 출력을 다음 층의 입력으로 사용
        return x  # 최종 출력 반환

    @property
    def weights(self):
        weights = []
        for layer in self.layers:
            weights += layer.weights  # 각 층의 가중치를 리스트에 추가
        return weights

# 예제 모델 정의
model = NaiveSequential([
    NaiveDense(input_size=28 * 28, output_size=512, activation=tf.nn.relu),
    NaiveDense(input_size=512, output_size=10, activation=tf.nn.softmax)
])

# 가중치 개수가 맞는지 확인
assert len(model.weights) == 4
```

```python
# NaiveDense 클래스 정의
class NaiveDense:
    def __init__(self, input_size, output_size, activation):
        self.input_size = input_size  # 입력 크기
        self.output_size = output_size  # 출력 크기
        self.activation = activation  # 활성화 함수

        # 가중치 초기화
        self.W = tf.Variable(tf.random.normal([input_size, output_size]), name='weights')
        # 편향 초기화
        self.b = tf.Variable(tf.zeros([output_size]), name='biases')

    def __call__(self, inputs):
        # 입력에 가중치를 곱하고 편향을 더한 후 활성화 함수를 적용하여 출력 계산
        return self.activation(tf.matmul(inputs, self.W) + self.b)

# 테스트용 NaiveDense 객체 생성
testNaiveDense_1 = NaiveDense(input_size=28 * 28, output_size=512, activation=tf.nn.relu)
testNaiveDense_2 = NaiveDense(input_size=512, output_size=10, activation=tf.nn.softmax)

# 각 NaiveDense 객체의 가중치 및 편향 출력
print("testNaiveDense_1.W : ", testNaiveDense_1.W)
print("testNaiveDense_1.b : ", testNaiveDense_1.b)
print("testNaiveDense_2.W : ", testNaiveDense_2.W)
print("testNaiveDense_2.b : ", testNaiveDense_2.b)

# 입력 데이터 생성 및 예측 과정
input_data_1 = tf.random.normal((1, 28 * 28))  # 입력 데이터 예시, 크기는 (배치 크기, 입력 특성 수)여야 함
print("input_data_1 : ", tf.shape(input_data_1), input_data_1)

# 첫 번째 NaiveDense 객체를 통한 출력 계산
output_1 = testNaiveDense_1(input_data_1)
print("output_1 :", tf.shape(output_1), output_1)

# 두 번째 NaiveDense 객체를 통한 출력 계산
output_2 = testNaiveDense_2(output_1)
print("output_2 :", tf.shape(output_2), output_2)
```

```python
import math

class BatchGenerator:
    def __init__(self, images, labels, batch_size=128):
        assert len(images) == len(labels)  # 이미지와 레이블의 길이가 같아야 함
        self.index = 0  # 현재 인덱스 초기화
        self.images = images  # 이미지 데이터
        self.labels = labels  # 레이블 데이터
        self.batch_size = batch_size  # 배치 크기
        self.num_batches = math.ceil(len(images) / batch_size)  # 전체 배치 개수 계산

    def next(self):
        # 현재 인덱스부터 배치 크기만큼의 이미지와 레이블을 가져옴
        images = self.images[self.index : self.index + self.batch_size]
        labels = self.labels[self.index : self.index + self.batch_size]
        self.index += self.batch_size  # 다음 배치를 위해 인덱스 업데이트
        return images, labels
```

   - 실제 레이블을 사용하여 예측의 손실 값 계산
   - 모델 가중치에 대한 손실의 그레디언트 계산
   - 그레디언트 반대 방향으로 가중치 이동

```python
import math

class BatchGenerator:
    def __init__(self, images, labels, batch_size=128):
        assert len(images) == len(labels)  # 이미지와 레이블의 길이가 같아야 함
        self.index = 0  # 현재 인덱스 초기화
        self.images = images  # 이미지 데이터
        self.labels = labels  # 레이블 데이터
        self.batch_size = batch_size  # 배치 크기
        self.num_batches = math.ceil(len(images) / batch_size)  # 전체 배치 개수 계산

    def next(self):
        # 현재 인덱스부터 배치 크기만큼의 이미지와 레이블을 가져옴
        images = self.images[self.index : self.index + self.batch_size]
        labels = self.labels[self.index : self.index + self.batch_size]
        self.index += self.batch_size  # 다음 배치를 위해 인덱스 업데이트
        return images, labels
```
